```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('outputDataT1.csv',header=None)
df.head()
```

```python
df.columns = ['instance','algo','seed','eps','horizon','REG']
```

```python
#df_group = df.groupby(['instance','algo','horizon']).mean()
```

```python
df_i1 = df[:1200]
df_i2 = df[1200:2400]
df_i3 = df[2400:]
```

```python
i_1_plot = df_i1.groupby(['algo','horizon']).mean()
```

```python
i_1_plot
```

```python
horizon = np.log([100,400,1600,6400,25600,102400])
```

```python
plt.figure(figsize=(12,8))
plt.plot(horizon,i_1_plot['REG'][0:6], label="epsilon-greedy")
plt.plot(horizon,i_1_plot['REG'][6:12],label="KL-UCB")
plt.plot(horizon,i_1_plot['REG'][12:18],label="Thompson-sampling")
plt.plot(horizon,i_1_plot['REG'][18:24],label="UCB")
plt.xlabel('log(horizon)')
plt.ylabel('Regret')
plt.title("Regret vs Horizon for Instance i-1.txt")
plt.legend()
plt.savefig("i1.png")
```
